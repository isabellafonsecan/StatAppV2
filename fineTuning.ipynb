```python
import os
from datasets import load_dataset

cache_dir = os.path.join("/home/onyxia/work", ".cache", "huggingface/datasets")
os.makedirs(cache_dir, exist_ok=True)

token = os.getenv("HF_TOKEN")

# Stream train split - downloads only what you use
ds_train = load_dataset(
    "marsianin500/Speech2Latex",
    split="train",
    streaming=True,  # KEY: Stream mode!
)

# Stream test split
ds_test = load_dataset(
    "marsianin500/Speech2Latex",
    split="test",
    streaming=True,
)

# Take only first N samples (downloads ONLY these)
train_subset = ds_train.take(5000)   # Downloads only 5000 samples
test_subset = ds_test.take(1000)     # Downloads only 1000 samples

print(f"✅ Train samples: 5000 (streamed)")
print(f"✅ Test samples: 1000 (streamed)")
print(f"📊 Combined dataset ready for fine-tuning")
```
